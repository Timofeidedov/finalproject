```python
def get_nth_page(year,month,day,horos):
    # скачиваем
    url = f'https://dommagii.com/horoscop/{year}-{month}-{day}'
    print(url)
    request = requests.get(url)
    print(request.status_code)
    if request.status_code == 200: 
      req = session.get(url, headers={'User-Agent': ua.random})
      page = req.text
      soup = BeautifulSoup(page, 'html.parser')
      text_horo=[]
      horo=soup.find_all('div',attrs={'class':'horo_text'})
      for smth in horo: 
        smth=smth.text
        text_horo.append(smth)
      for smth in text_horo:
        print(smth)
        horos.append(smth)
```

```python
nyear=2010
nmonth=9
nday=1
horos=[]
while nyear<2019:
  year=str(nyear)
  while nmonth<13:
    if nmonth<10:
      month='0'+str(nmonth)
    else:
      month=str(nmonth)
    while nday<32:
      if nday<10:
        day='0'+str(nday)
      else:
        day=str(nday)
      get_nth_page(year,month,day,horos)
      nday+=1
    nmonth+=1
    nday=1
  nyear+=1
  nmonth=1
  nday=1
```
